```python
import pandas as pd
import os
```

```python
# Diretório dos arquivos Excel
diretorio = r"C:\VilarLearn\Source\Travel"

# Lista para armazenar os DataFrames dos arquivos
dados = []

# Lista para armazenar os nomes dos arquivos
nomes_arquivos = []

# Iterar sobre os arquivos Excel no diretório
for arquivo in os.listdir(diretorio):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        dados_arquivo = pd.read_excel(caminho_arquivo)
        
        # Adicionar coluna com o nome do arquivo
        dados_arquivo['Nome_Arquivo'] = arquivo
        
        dados.append(dados_arquivo)
        nomes_arquivos.extend([arquivo] * len(dados_arquivo))

# Consolidar os dados em um único DataFrame
consolidado = pd.concat(dados, ignore_index=True)

# Exibir o DataFrame consolidado
consolidado
```

```python
from sqlalchemy import create_engine

# Configurações da conexão com o Azure SQL
server = 'seu_server.database.windows.net'
database = 'seu_banco_de_dados'
username = 'Seu_usuario'
password = 'Sua_senha'
driver = 'ODBC Driver 17 for SQL Server'

# String de conexão
conn_str = f"mssql+pyodbc://{username}:{password}@{server}/{database}?driver={driver}"


engine = create_engine(conn_str)

# Nome da tabela no Azure SQL
nome_tabela = 'Viagens'

# Enviar DataFrame para o Azure SQL
consolidado.to_sql(nome_tabela, con=engine, if_exists='replace', index=False)

# Fechar a conexão
engine.dispose()
```
