**Adicionando nova coluna na tabela**

```python
#Criando nova coluna 
tb['Nova Coluna'] = 'Não Informado'
tb.head(5)
```

```python
#retorna a data do dia corrente
from datetime import date

data_atual = date.today()
print(data_atual)
```

```python
#adiciona coluna da data atual no dataframe
tb['data_atualizacao'] = data_atual
tb
```

```python
#formata a data em string - DDMMYYYY
data_em_texto = data_atual.strftime('%d/%m/%Y')
data_em_texto
```

```python
from datetime import datetime

data_e_hora_atuais = datetime.now()
data_e_hora_em_texto = data_e_hora_atuais.strftime('%d/%m/%Y')

print(data_e_hora_em_texto)
```

```python
#resolve problema de fuso horário no horário da data e hora
from pandas.core.arrays.datetimelike import timedelta
from datetime import datetime, timezone

diferenca = timedelta(hours=-3)
print(diferenca)

fuso_horario = timezone(diferenca)
print(fuso_horario)

data_e_hora_bsb = data_e_hora_atuais.astimezone(fuso_horario)
data_e_hora_bsb_em_texto = data_e_hora_bsb.strftime('%d/%m/%Y %H:%M')

print(data_e_hora_bsb_em_texto)
```

```python
#Cria coluna com condicional
tb['Dia_Semana'] = tb.day.apply(lambda x: 'Sábado' if 'Sat' in x 
                                       else ('Domingo' if 'Sun' in x 
                                             else ('Quinta-feira' if 'Thur' in x
                                              else ('Sexta-feira' if 'Friday' in x 
                                                   else 'NI'))))
tb.tail(10)
```

**Exclusão de coluna**

```python
#excluir colunas
tb.drop(columns=['data_atualizacao','Dia_Semana'], 
                inplace=True)
tb
```

**Renomear coluna e campo**

```python
#renomeia o nome da coluna
tb.rename(columns={'day':'dia',
                   'sex':'sexo'},
          inplace=True)
tb
```

```python
#renomeia o nome do registro da coluna

tb['sexo'] = tb['sexo'].map({'Male':'Masculino',
                             'Female':'Feminimo'},
                             na_action=None)

tb
```

**Utilizando INNER JOIN**

```python
#inner join
import pandas as pd
  
# Criando um dataframe A
a = pd.DataFrame()
  
# Criando um dicionário com os dados para tabela A
dados = {'id': [1, 2, 10, 12,15],
     'val1': ['a', 'b', 'c', 'd','e']}
  
a = pd.DataFrame(dados)

#############################################################
  
# Criando um dataframe B
b = pd.DataFrame()
  
# Criando um dicionário com os dados para tabela B
dados = {'id': [1, 2, 9, 8,15],
     'val1': ['p', 'q', 'r', 's','x']}
b = pd.DataFrame(dados)
  
# inner join
df = pd.merge(a, b, on='id', how='inner')
  
# display dataframe
df
```
