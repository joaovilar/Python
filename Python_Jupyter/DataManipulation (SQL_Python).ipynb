```python
#importando as bibliotecas que serão utilizadas na conexão e manipulação com sql

from sqlalchemy import create_engine
import pymysql
import pandas as pd
import numpy as np
```

```python
#string de conexão com o MySQL- //usuario:senha@servidor/NomeDoBancoDeDados

conexao = create_engine('mysql+pymysql://root:root@localhost/teste')
```

```python
##Consultando os dados do MySQL e guardando em um dataframe usando o pandas
#Tabela de viagens
df = pd.read_sql('SELECT * FROM viagens', con=conexao)
df
```

## AND

#### SQL

```python
df_and = pd.read_sql("""
SELECT * FROM viagens 
WHERE Continente ='Europa' AND Preço_médio_do_hotel >= 500.0 """, con=conexao)
df_and
```

#### PYTHON

```python
df[(df['Continente'] == 'Europa') & (df['Preço_médio_do_hotel'] >= 500.0)].sort_values('Preço_médio_do_hotel')
```

## OR

#### SQL

```python
df_or = pd.read_sql("""
SELECT * FROM viagens 
WHERE Continente ='Europa' OR Continente = 'América Central' order by Preço_médio_do_hotel desc""", con=conexao)
df_or
```

#### PYTHON

```python
df[(df['Continente'] =='Europa') | (df['Continente'] =='América Central')].sort_values('Preço_médio_do_hotel', ascending=False)
```
