```python
SERVER_NAME = 'localhost\SQLEXPRESS01'
DATABASE_NAME = 'BD_VENDAS'
```

```python
#conectando através do drive odbc 

conn = pypyodbc.connect("""
                  DRIVER={{SQL Server}};
                  SERVER={0};
                  DATABASE={1};
               
                  TrustServerCertificate=no;""".format(SERVER_NAME, DATABASE_NAME)

)
```

```python
sql_query = """

SELECT *
  FROM [BD_VENDAS].[dbo].[TBProdutos]
  
"""
```

```python
mssql = pd.read_sql(sql_query,conn)

mssql
```

```python
#renomeando um registro na coluna
mssql['cidade'] = mssql['cidade'].map({'SÃ£o Paulo':'São Paulo'},
                                na_action=None)
```

## 4 - Conectando com Banco de dados MongoDB

Instalando bibiotecas importantes para o uso na conexão

```python
!pip install pymongo
```

```python
# importando as bibliotecas para o uso com o mongodb

import pymongo

from pymongo import MongoClient
```

```python
#conexão

client = MongoClient("mongodb://localhost", 27017)
```

```python
#Informação do nome do banco de dados

db = client['AprendendoMongo']

db
```

```python
#listando os bancos de dados

client.list_database_names()
```

```python
#listando as collection que possuo

collect_names = db.list_collection_names()

collect_names
```

```python
# retorna a quantidade de documentos para essa collection

Disciplinas = db["Disciplinas"]

Disciplinas.count_documents({})
```

```python
# Listando todos os documentos da minha collection
Disciplinas = db["Disciplinas"]

for Disciplinas in Disciplinas.find():
    print(Disciplinas)
```
