```python
from google.colab import files
import pandas as pd
import io

# Carregar o arquivo / upload files
uploaded = files.upload()

# Ler o conteúdo do arquivo / Read the contents of the file
for filename, content in uploaded.items():
    # Use o nome do arquivo para criar o DataFrame / Use the filename to create the DataFrame
    df = pd.read_csv(io.BytesIO(content))

# Exiba o DataFrame / display the dataframe
df
```

```python
#cria coluna no dataframe com a data do dia corrente
from datetime import datetime
df['data_atual'] = datetime.now().date()

df
```
